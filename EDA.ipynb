```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
sns.set(style="whitegrid")
%matplotlib inline

# Load dataset (uploaded train.csv)
df = pd.read_csv('train.csv')

# Display basic info
df.info()

# Display descriptive statistics
df.describe()

# Value counts for important categorical variables
print("Survived counts:\n", df['Survived'].value_counts())
print("Pclass counts:\n", df['Pclass'].value_counts())
print("Sex counts:\n", df['Sex'].value_counts())
print("Embarked counts:\n", df['Embarked'].value_counts())

# Pairplot (only on numeric features)
numeric_cols = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
sns.pairplot(df[numeric_cols], hue='Survived')
plt.suptitle('Pairplot of Numeric Features', y=1.02)
plt.show()

# Heatmap of correlations
plt.figure(figsize=(10,8))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Histograms
numeric_features = ['Age', 'Fare', 'SibSp', 'Parch']
for col in numeric_features:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()

# Boxplots
plt.figure(figsize=(6,4))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Boxplot of Fare by Pclass')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Boxplot of Age by Survival')
plt.show()

# Scatterplots
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Scatterplot of Age vs Fare (colored by Survival)')
plt.show()


```
